# Checkpoint 1: Data Cleaning 1

- Load the companies and rounds data into two data frames and name them companies and rounds2 respectively.

```python
# Read companies.txt and create a data frame companies. 
# use mbcs to encode the txt data file
companies = pd.read_csv('./companies.txt',sep="\t",encoding='iso-8859-1')
```

```python
# decode the permalink
companies['permalink'] = companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
```

1. Spark Funds wants to see the top nine countries which have received the highest total funding (across ALL sectors for the chosen investment type)

2. For the chosen investment type, make a data frame named top9 with the top nine countries (based on the total investment amount each country has received)

```python
# update country code null with Missing

master_frame['country_code'].fillna('MISSING',inplace=True)

# check the country_code for MISSING
master_frame['country_code'].head(5)
```

```python
# the chosen investment for spark fund is venture. 
# create a dataframe to store the country_code and total fund raised for the investement "venture"
# Exclude the countries with no country code from the top9 by excluding MISSING country_code

tot_fund_country = master_frame.loc[(master_frame['funding_round_type']=='venture')  & (master_frame['country_code']!='MISSING')].groupby(['country_code']).sum()

# sort the tot_fun_country data frame descending and choose top9 countries with higher investments.
tot_fund_country = tot_fund_country.sort_values(by='raised_amount_usd', ascending=False)

# store the top 9 countries with highest total funding into top 9 dataframe

top9=tot_fund_country.head(9)

# check the top 9 countries by highest investment from top9 dataframe 
top9
```

```python
# from the analysis the top english speaking country 

# from the provided pdf based on the above top9 dataframe create english_speaking_df 

english_speaking_df = pd.DataFrame({'country_code':['USA','CHN','GBR','IND','CAN','FRA','ISR','DEU','JPN'],
                                        'country_name':['United States','China','United Kingdom','India','Canada','France','Israel','Germany','Japan'],
                                   'english_speaking':['Y','N','Y','Y','Y','N','N','N','N']})
english_speaking_df

```

```python
# Top3 english speaking countries based on investment amount 

top_offical_speaking_merge = pd.merge(top9,english_speaking_df,on='country_code')

top_offical_speaking_merge = top_offical_speaking_merge.loc[top_offical_speaking_merge['english_speaking']=='Y']
top3_english_speaking_ctry = top_offical_speaking_merge.head(3)

top3_english_speaking_ctry.index=range(3)
top3_english_speaking_ctry.index = top3_english_speaking_ctry.index+1

top3_english_speaking_ctry

```

# Checkpoint 5: Sector Analysis 2

Based on initial analysis the Top 3 English speaking countries wiht high investments are 

1. USA 
2. GBR
3. IND

Create three separate data frames __D1, D2 and D3 for each of the three countries__ containing the observations of funding type FT falling within the __5-15 million USD__ range. The three data frames should contain:

- All the columns of the master_frame along with the primary sector and the main sector

- The total number (or count) of investments for each main sector in a separate column

- The total amount invested in each main sector in a separate column

Using the three data frames, you can calculate the total number and amount of investments in each main sector.

```python
# data frame D1 for country_code 'USA'

D1 = master_frame.loc[(master_frame['country_code']=='USA') & (master_frame['funding_round_type']=='venture')]

# Filter 5-15 million USD range

D1 = D1.loc[(D1['raised_amount_usd']/1000000>=5) & (D1['raised_amount_usd']/1000000<=15)]

# add separate columns to store the count and sum across main_sector

D1['total_investment_count'] = D1.main_sector.map(D1.groupby('main_sector')['raised_amount_usd'].count())

D1['total_investment_Amt'] = D1.main_sector.map(D1.groupby('main_sector')['raised_amount_usd'].sum())

D1.head()
```

```python
# check the values in D1 for sum
D1.groupby(['main_sector'])['raised_amount_usd'].sum().sort_values(ascending=False)
```

```python
# check the values in D1 for count
D1.groupby(['main_sector'])['raised_amount_usd'].count().sort_values(ascending=False)
```

```python
# data frame D2 for country_code 'GBR'

D2 = master_frame.loc[(master_frame['country_code']=='GBR') & (master_frame['funding_round_type']=='venture')]

# Filter 5-15 million USD range

D2 = D2.loc[(D2['raised_amount_usd']/1000000>=5) & (D2['raised_amount_usd']/1000000<=15)]

# add separate columns to store the count and sum across main_sector

D2['total_investment_count'] = D2.main_sector.map(D2.groupby('main_sector')['raised_amount_usd'].count())

D2['total_investment_Amt'] = D2.main_sector.map(D2.groupby('main_sector')['raised_amount_usd'].sum())

D2.head()
```

```python
# data frame D3 for country_code 'IND'

D3 = master_frame.loc[(master_frame['country_code']=='IND') & (master_frame['funding_round_type']=='venture')]

# Filter 5-15 million USD range

D3 = D3.loc[(D3['raised_amount_usd']/1000000>=5) & (D3['raised_amount_usd']/1000000<=15)]

# add separate columns to store the count and sum across main_sector

D3['total_investment_count'] = D3.main_sector.map(D3.groupby('main_sector')['raised_amount_usd'].count())

D3['total_investment_Amt'] = D3.main_sector.map(D3.groupby('main_sector')['raised_amount_usd'].sum())

D3.head()
```

```python
# concat the three data frames D1, D2 and D3 into 1 dataframe so it will be easy for the analysis 
top3_country_df = pd.concat([D1,D2,D3], ignore_index=True)
```

```python
# Total number of investment count for countries USA, GBR and IND
# if there is 0 inverstment remove it from count

top3_country_df.loc[top3_country_df['raised_amount_usd'] !=0].groupby('country_code')['raised_amount_usd'].count().sort_values(ascending=False)
```
